```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
style.use('ggplot')
# style.use('fivethirtyeight')
```

```python
df = pd.read_excel('potentiel_energetico.xlsx', sheetname='tabla_es', skiprows=2)
df.rename(inplace= True, columns={
    'Unnamed: 0': 'Categoria',
    'SUSTRATO': 'Sustrato',
    'Producción biogás litro/kg oMS': 'Producción de Bíogas l/kg oMS', 
    'Produccion electricidad [1] kWh/TMB': 'Generación Electrica kWh/T MB'})
df.drop(df.columns[[2, 3, 5]], axis=1, inplace=True)
df.drop(df.index[[111, 112]], inplace=True)
df.set_index('Sustrato', inplace=True)
```

```python
df.head()
```

```python
ax_one = df.plot(kind='bar', figsize=(50,25))
ax_one.plot()
```

```python
ax_two = df.plot(kind='bar', figsize=(20,10))
ax_two.plot()
```

```python
ax_three = df.groupby('Categoria')
ax_three.plot(kind='bar', figsize=(12,8))
```

```python
ax_four = df.groupby('Categoria').mean()
ax_four.plot(kind='bar', figsize=(15,10))
```

```python
fig_one = ax_one.get_figure()
fig_one.savefig('plot_one.png')

fig_two = ax_two.get_figure()
fig_two.savefig('plot_two.png')

# fig_three = ax_three.get_figure()
# fig_three.savefig('plot_three.png')

# fig_four = ax_four.get_figure()
# fig_four.savefig('plot_four.png')
```
