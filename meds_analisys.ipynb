```python
# Una manera de ver los datos un poco mas crudos, pero se desordena esta vaina.
pd.pivot_table(df, index=["Identificacion Paciente"], values=["Codigo Atc"], aggfunc=pd.Series.nunique, fill_value=0, columns=["Fecha Entrega Formula"])
```

```python
# Veamos como estan esos rangos de edad para agregarlos
print("Edad minima:", df["Edad."].min(), "\t", "Edad maxima:", df["Edad."].max())
```

```python
# y Pues voila, rangos de edad (por lo menos segun lo que dice el dato anterior)
bins = [60, 70, 80, 90, 100]
labels = ["60 a 70", "70 a 80", "90 a 100", "100 a 110"]
df["Rango de Edad"] = pd.cut(df["Edad."], bins=bins, labels=labels)
# Listo, ahora veamos cuantos medicamentos por mes por año agrupado por rango de edades.
pd.pivot_table(df, index=["Rango de Edad", "Identificacion Paciente", "Año", "Mes."], values=["Codigo Atc"], aggfunc=pd.Series.nunique, fill_value=0)
```

```python
df = df[["Año", "Mes.", "Fecha Entrega Formula", "Identificacion Paciente", "Edad.", "Codigo Atc", "Rango de Edad"]]
df.head()
```

```python
df.set_index("Fecha Entrega Formula").groupby(["Identificacion Paciente", "Rango de Edad"]).resample('M').sum()
```

```python
intermediate_df = df.groupby(['Identificacion Paciente']).agg({'Codigo Atc': pd.Series.nunique, 'Fecha Entrega Formula': pd.Series.nunique})
df = df[df['Identificacion Paciente'].isin(intermediate_df[intermediate_df['Codigo Atc'] > 4].index)]
df = df[df['Identificacion Paciente'].isin(intermediate_df[intermediate_df['Fecha Entrega Formula'] > 1].index)]
df.sort_values(by=['Identificacion Paciente', 'Fecha Entrega Formula'])
```
